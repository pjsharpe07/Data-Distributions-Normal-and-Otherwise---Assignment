```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
print('Import Successful')
```

```python
normal_dist = np.random.normal(0, 10, 100)

plt.hist(normal_dist, color='c')

plt.axvline(normal_dist.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(normal_dist.mean() + normal_dist.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(normal_dist.mean() - normal_dist.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('Normal Distribution')
plt.show()
```

```python
 hyper_geo = np.random.hypergeometric(100, 2, 10, 100)
    
plt.hist(hyper_geo, color='c')

plt.axvline(hyper_geo.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(hyper_geo.mean() + hyper_geo.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(hyper_geo.mean() - hyper_geo.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('Hyper Geometric Distribution')
plt.show()
```

```python
gumbel = np.random.gumbel(0, 0.1, 100)

plt.hist(gumbel, color='c')

plt.axvline(gumbel.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(gumbel.mean() + gumbel.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(gumbel.mean() - gumbel.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('Gumbel Distribution')
plt.show()
```

```python
rand1 = np.random.normal(5, 0.5, 100)
rand2 = np.random.normal(10, 1, 100)

combined = rand1 + rand2

plt.hist(combined, color='c')

plt.axvline(combined.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(combined.mean() + combined.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(combined.mean() - combined.std(), color='b', linestyle='dashed', linewidth=2)

plt.title('Gumbel Distribution')
plt.show()
```
